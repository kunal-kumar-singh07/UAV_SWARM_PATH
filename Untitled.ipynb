```python
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

random.seed(42)
np.random.seed(42)

GRID_W = 20
GRID_H = 10
N_CELLS = GRID_W * GRID_H
N_UAV = 15
BASE_LOCATIONS = [(0, 0), (0, GRID_H-1), (GRID_W-1, 0), (GRID_W-1, GRID_H-1)]
UAV_BASES = [BASE_LOCATIONS[i % len(BASE_LOCATIONS)] for i in range(N_UAV)]

ndvi = np.clip(np.random.normal(loc=0.6, scale=0.18, size=(GRID_H, GRID_W)), 0.0, 1.0)

for cx, cy, r, depth in [(8,4,3,-0.4), (15,7,2,-0.35)]:
    for y in range(GRID_H):
        for x in range(GRID_W):
            d = math.hypot(x-cx, y-cy)
            if d <= r:
                ndvi[y,x] = np.clip(ndvi[y,x] + depth*(1 - d/(r+1)), 0.0, 1.0)

cells = []
for y in range(GRID_H):
    for x in range(GRID_W):
        idx = y*GRID_W + x
        cells.append({"idx": idx, "x": x+0.5, "y": y+0.5, "ndvi": float(ndvi[y,x])})

cells_df = pd.DataFrame(cells)

restricted_mask = np.zeros((GRID_H, GRID_W), dtype=bool)
for cx, cy, r in [(5,2,2), (12,6,2)]:
    for y in range(GRID_H):
        for x in range(GRID_W):
            if math.hypot(x-cx, y-cy) <= r:
                restricted_mask[y,x] = True

cells_df['restricted'] = [restricted_mask[int(i//GRID_W), int(i%GRID_W)] for i in cells_df['idx']]

threshold = np.quantile(cells_df['ndvi'], 0.25)
hp_indices = cells_df[cells_df['ndvi'] <= threshold].sample(frac=0.6, random_state=42)['idx'].tolist()
cells_df['double_scan'] = cells_df['idx'].apply(lambda i: i in hp_indices)

uav_batteries = [random.uniform(0.6, 1.0) for _ in range(N_UAV)]
ENERGY_PER_DIST = 0.0015
ENERGY_PER_SCAN = 0.0008
SCAN_TIME = 0.25
SPEED = 1.2
SEPARATION = 1.5
DOUBLE_SCAN_ADDITIONAL = 1

def dist(a,b):
    return math.hypot(a[0]-b[0], a[1]-b[1])

coord = {row.idx: (row.x, row.y) for row in cells_df.itertuples()}

POP_SIZE = 120
GENERATIONS = 120
MUT_RATE = 0.02
TOURNAMENT = 3
ELITE = 4

def random_chromosome():
    arr = np.zeros(N_CELLS, dtype=int)
    for i in range(N_CELLS):
        p = coord[i]
        dists = [dist(p, UAV_BASES[u]) for u in range(N_UAV)]
        sorted_u = np.argsort(dists)
        arr[i] = int(random.choice(sorted_u[:3].tolist()))
    return arr

def decode_routes(assignment):
    routes = {u: [] for u in range(N_UAV)}
    for i,u in enumerate(assignment):
        routes[u].append(i)
    ordered = {}
    for u in range(N_UAV):
        seq = routes[u]
        if not seq:
            ordered[u] = []
            continue
        current = UAV_BASES[u]
        remaining = set(seq)
        order = []
        while remaining:
            nxt = min(remaining, key=lambda idx: dist(current, coord[idx]))
            order.append(nxt)
            remaining.remove(nxt)
            current = coord[nxt]
        ordered[u] = order
    return ordered

def evaluate(assignment):
    routes = decode_routes(assignment)
    total_time = 0
    total_energy = 0
    penalty = 0
    events = []
    for u in range(N_UAV):
        route = routes[u]
        if not route:
            continue
        t = 0
        cur = UAV_BASES[u]
        battery = uav_batteries[u]
        for c in route:
            d = dist(cur, coord[c])
            t += d / SPEED
            scans = 2 if cells_df.loc[c,'double_scan'] else 1
            t += SCAN_TIME * scans
            e = d * ENERGY_PER_DIST + scans * ENERGY_PER_SCAN
            battery -= e
            total_energy += e
            if cells_df.loc[c,'restricted']:
                penalty += 5
            events.append((u, c, t, coord[c]))
            cur = coord[c]
        dback = dist(cur, UAV_BASES[u])
        t += dback / SPEED
        e = dback * ENERGY_PER_DIST
        battery -= e
        total_energy += e
        if battery < 0.05:
            penalty += 8 + abs(battery)*30
        total_time = max(total_time, t)
    events = sorted(events, key=lambda x: x[2])
    for i in range(len(events)):
        u1, c1, t1, p1 = events[i]
        for j in range(i+1, min(i+40, len(events))):
            u2, c2, t2, p2 = events[j]
            if u1 == u2:
                continue
            if abs(t1 - t2) > 0.6:
                break
            if dist(p1,p2) < SEPARATION:
                penalty += 4
    return total_time + 30*total_energy + penalty

def tournament_select(pop, fits):
    i = random.randrange(len(pop))
    best = i
    for _ in range(TOURNAMENT-1):
        j = random.randrange(len(pop))
        if fits[j] < fits[best]:
            best = j
    return pop[best].copy()

def crossover(a,b):
    cut = random.randint(1, N_CELLS-2)
    child = a.copy()
    child[:cut] = a[:cut]
    child[cut:] = b[cut:]
    return child

def mutate(chrom):
    for i in range(N_CELLS):
        if random.random() < MUT_RATE:
            p = coord[i]
            dists = [dist(p, UAV_BASES[u]) for u in range(N_UAV)]
            sorted_u = np.argsort(dists)
            chrom[i] = int(random.choice(sorted_u[:3].tolist()))
    return chrom

population = [random_chromosome() for _ in range(POP_SIZE)]
best_fit = float('inf')
best_chrom = None

for gen in range(GENERATIONS):
    fits = [evaluate(ind) for ind in population]
    idx = np.argsort(fits)
    newpop = [population[i].copy() for i in idx[:ELITE]]
    if fits[idx[0]] < best_fit:
        best_fit = fits[idx[0]]
        best_chrom = population[idx[0]].copy()
    while len(newpop) < POP_SIZE:
        p1 = tournament_select(population, fits)
        p2 = tournament_select(population, fits)
        child = crossover(p1,p2)
        child = mutate(child)
        newpop.append(child)
    population = newpop

best_routes = decode_routes(best_chrom)
score = evaluate(best_chrom)

alloc = []
for i in range(N_CELLS):
    alloc.append({
        "cell_idx": i,
        "x": coord[i][0],
        "y": coord[i][1],
        "ndvi": cells_df.loc[i,'ndvi'],
        "restricted": cells_df.loc[i,'restricted'],
        "double_scan": cells_df.loc[i,'double_scan'],
        "assigned_uav": int(best_chrom[i])
    })

alloc_df = pd.DataFrame(alloc)
display(alloc_df)

uav_summary = []
for u in range(N_UAV):
    route = best_routes[u]
    if not route:
        uav_summary.append({"uav":u,"base":UAV_BASES[u],"assigned_cells":0})
        continue
    cur = UAV_BASES[u]
    t = 0
    e = 0
    for c in route:
        d = dist(cur, coord[c])
        t += d/SPEED + SCAN_TIME * (2 if cells_df.loc[c,'double_scan'] else 1)
        e += d*ENERGY_PER_DIST + (2 if cells_df.loc[c,'double_scan'] else 1)*ENERGY_PER_SCAN
        cur = coord[c]
    dback = dist(cur, UAV_BASES[u])
    t += dback/SPEED
    e += dback*ENERGY_PER_DIST
    uav_summary.append({
        "uav":u,
        "base":UAV_BASES[u],
        "assigned_cells":len(route),
        "time_min":round(t,3),
        "energy_used":round(e,5),
        "battery_start":round(uav_batteries[u],3)
    })

uav_summary_df = pd.DataFrame(uav_summary)
display(uav_summary_df)

plt.figure(figsize=(8,4))
plt.imshow(ndvi, origin='lower', extent=[0, GRID_W, 0, GRID_H], aspect='auto')
for i,row in alloc_df.iterrows():
    plt.scatter(row['x'], row['y'], s=12)
rx,ry = np.where(restricted_mask.T==True)
plt.scatter(rx+0.5, ry+0.5, s=40, marker='x')
plt.title("Farm NDVI with UAV Assignments")
plt.tight_layout()
plt.show()

print("Final Fitness:", score)
print("Sample Routes:")
for u in range(min(6,N_UAV)):
    print("UAV",u,":",best_routes[u][:10])
```
